```python
''' selecting malnutrition in the series and stunting to get data of malnourished children '''

df_malnutr=df_sdg[df_sdg['Series'].str.contains('malnutrition') | df_sdg['Series'].str.contains('stunting')]

#drop some unimportant columns

df_malnutr=df_malnutr.drop(columns=['Unit_Code','Ref_Area','Province','District','Urbanization','Education ',2000,2001,2002,2003,2004])

#transpose the dataset

df_malnutr=df_malnutr.T

#reduce the columns and select the column names

df_malnutr.columns=df_malnutr.iloc[3]
df_malnutr=df_malnutr[5:]

#stunted,wasted and overweight malnutrition types
#stunting
'''
here, we are going to make three different datasets for each malnutrition type

1. stunting : height by age

2. wasting : weight for height

3. overweight:


'''
#stunting

df_malnutr_stunted=df_malnutr['Proportion of children moderately or severely stunted']

df_malnutr_stunted.columns=df_malnutr_stunted.iloc[0]
df_malnutr_stunted=df_malnutr_stunted[1:]
df_malnutr_stunted=df_malnutr_stunted.fillna(method='ffill')
df_malnutr_stunted['indicator']='stunting'
#print(df_malnutr_stunted)

#wasting

df_malnutr_wasting=df_malnutr['Proportion of children moderately or severely wasted']
df_malnutr_wasting.columns=df_malnutr_wasting.iloc[0]
df_malnutr_wasting=df_malnutr_wasting[1:]
df_malnutr_wasting=df_malnutr_wasting.fillna(method='ffill')
df_malnutr_wasting['indicator']='wasting'
#print(df_malnutr_wasting)

#overweight

df_malnutr_overweight=df_malnutr['Proportion of children moderately or severely overweight']
df_malnutr_overweight.columns=df_malnutr_overweight.iloc[0]
df_malnutr_overweight=df_malnutr_overweight[6:]
df_malnutr_overweight=df_malnutr_overweight.fillna(method='ffill')
df_malnutr_overweight['indicator']='overweight'
#print(df_malnutr_overweight)
#combines all the sub datasets to make malnutrition dataset
malnutrition=pd.concat([df_malnutr_stunted,df_malnutr_wasting,df_malnutr_overweight])
#malnutrition
```

```python
''' 
selecting health data from the dataset

1. Proportion of births attended by skilled health personnel
2. Under-five mortality rate
3. Neonatal mortality rate

'''
#get health children related data
df_health=df_sdg[df_sdg['Series'].str.contains('births attended by skilled health personnel') | df_sdg['Series'].str.contains('Under-five') | df_sdg['Series'].str.contains('Neonatal') | df_sdg['Series'].str.contains('new HIV')]

#drop unimportant columns
df_health=df_health.drop(columns=['Unit_Code','Ref_Area','Data Source','Education ','District','Description','Age'])
#transpose the dataset
df_health=df_health.T

#changing the columns headers of dataset
df_health.columns=df_health.iloc[1]
df_health=df_health[2:]

# 1. Proportion of births attended by skilled health personnel

df_health_birth_attended_skilled=df_health['Proportion of births attended by skilled health personnel']

#remove sex row from the dataset as it has no value in Proportion of births attended by skilled health personnel
df_health_birth_attended_skilled=df_health_birth_attended_skilled.drop(index='Sex')

#select data related to residence for Proportion of births attended by skilled health personnel

'''
for this task, we select all rows and consider only the first four columns
which includes information on residence
we fill forward because data is updated when new survey results is out

source: DHS (Demographic and Health Survey)
'''
df_health_birth_attended_res=df_health_birth_attended_skilled.iloc[::,:3]
df_health_birth_attended_res=df_health_birth_attended_res.drop(index='Province')
df_health_birth_attended_res.columns=df_health_birth_attended_res.iloc[0]
df_health_birth_attended_res=df_health_birth_attended_res[1:]
df_health_birth_attended_res=df_health_birth_attended_res.fillna(method='ffill')
df_health_birth_attended_res=df_health_birth_attended_res.dropna(axis = 0, how ='all')
df_health_birth_attended_res = df_health_birth_attended_res.reset_index()
df_health_birth_attended_res=df_health_birth_attended_res.melt(id_vars='index')
#df_health_birth_attended_res

#select data related to provinces

'''
for this task, we select all rows and consider only the columns where there is province data
we delete data related to location

we make the provinces column as the header 
we do fill forward method to replace null values
rows with null values are dropped out

'''
df_health_birth_attended_res=df_health_birth_attended_skilled.iloc[::,:3]
df_health_birth_attended_res=df_health_birth_attended_res.drop(index='Province')
df_health_birth_attended_res.columns=df_health_birth_attended_res.iloc[0]
df_health_birth_attended_res=df_health_birth_attended_res[1:]
df_health_birth_attended_res=df_health_birth_attended_res.fillna(method='ffill')
df_health_birth_attended_res=df_health_birth_attended_res.dropna(axis = 0, how ='all')
df_health_birth_attended_res = df_health_birth_attended_res.reset_index()
df_health_birth_attended_res=df_health_birth_attended_res.melt(id_vars='index')
#df_health_birth_attended_res


#2. Under five mortality rate
'''
create dataframe contained underfive mortality data
remove unimportant columns
do manipulations: fill forward, transpose rows and columns
'''
df_health_underfive=df_sdg[df_sdg['Series']=='Under_five mortality rate']
df_health_underfive=df_health_underfive.drop(columns=['Unit_Code','Ref_Area','Education ','Data Source','Description','Indicator','Age'])
df_health_underfive=df_health_underfive.T

#create table under five mortality rate (per 1000) by provinces
#remove unimportant columns 
df_health_underfive.columns=df_health_underfive.iloc[1]
df_health_underfive_province=df_health_underfive[3:]
df_health_underfive_province=df_health_underfive_province.drop(index=['Urbanization','Sex'])
df_health_underfive_province=df_health_underfive_province.iloc[::,5:]
df_health_underfive_province=df_health_underfive_province.fillna(method='ffill')
df_health_underfive_province=df_health_underfive_province.dropna()
df_health_underfive_province=df_health_underfive_province.reset_index()
df_health_underfive_province=df_health_underfive_province.melt(id_vars='index')
#df_health_underfive_province

#create a table of underfive mortality by residence

df_health_underfive.columns=df_health_underfive.iloc[3]
df_health_underfive_res=df_health_underfive.iloc[::,:5]
df_health_underfive_res=df_health_underfive_res.drop(index=['Series','Province','District','Urbanization','Sex'])
df_health_underfive_res=df_health_underfive_res[['Urban','Rural']]
df_health_underfive_res=df_health_underfive_res.fillna(method='ffill')
df_health_underfive_res=df_health_underfive_res.dropna()
df_health_underfive_res=df_health_underfive_res.reset_index()
df_health_underfive_res=df_health_underfive_res.melt(id_vars='index')
#df_health_underfive_res

#create a table of underfive mortality by sex

df_health_underfive.columns=df_health_underfive.iloc[4]
df_health_underfive_sex=df_health_underfive.iloc[::,:3]
df_health_underfive_sex=df_health_underfive_sex.drop(index=['Series','Province','District','Urbanization','Sex'])
df_health_underfive_sex=df_health_underfive_sex.fillna(method='ffill')
df_health_underfive_sex=df_health_underfive_sex.dropna(axis=0,how='all')
df_health_underfive_sex=df_health_underfive_sex.reset_index()
df_health_underfive_sex=df_health_underfive_sex.melt(id_vars='index')
#df_health_underfive_sex

#3. Neonatal mortality rate (per 1000 live birth)
df_health_neonatal=df_health['Neonatal mortality rate']
df_health_neonatal.columns=df_health_neonatal.iloc[1]
df_health_neonatal_res=df_health_neonatal.iloc[::,:3]
df_health_neonatal_res=df_health_neonatal_res.drop(index=['Province','Urbanization','Sex'])
df_health_neonatal_res=df_health_neonatal_res.fillna(method='ffill')
df_health_neonatal_res=df_health_neonatal_res.dropna(axis=0,how='all')
df_health_neonatal_res=df_health_neonatal_res.reset_index()
df_health_neonatal_res=df_health_neonatal_res.melt(id_vars='index')
#df_health_neonatal_res

#Neonatal mortality rate by sex
#Neonatal mortality rate by sex
df_health_neonatal.columns=df_health_neonatal.iloc[2]
df_health_neonatal_sex=df_health_neonatal.iloc[::,3:]
df_health_neonatal_sex=df_health_neonatal_sex.drop(index=['Province','Urbanization','Sex'])
df_health_neonatal_sex=df_health_neonatal_sex.fillna(method='ffill')
df_health_neonatal_sex=df_health_neonatal_sex.dropna(axis=0,how='all')
df_health_neonatal_sex=df_health_neonatal_sex.reset_index()
df_health_neonatal_sex=df_health_neonatal_sex.melt(id_vars='index')
#df_health_neonatal_sex

#Adolescent birth rate by residence

df_health_adolescent_res=df_sdg[df_sdg['Series'].str.contains('Adolescent')]
df_health_adolescent_res=df_health_adolescent_res.drop(columns=['Indicator','Unit_Code','Ref_Area','Province','District','Education ','Data Source'])
df_health_adolescent_res=df_health_adolescent_res.T
df_health_adolescent_res.columns=df_health_adolescent_res.iloc[1]
df_health_adolescent_res=df_health_adolescent_res.drop(index=['Series','Urbanization','Description','Age','Sex'])
df_health_adolescent_res=df_health_adolescent_res.fillna(method='ffill')
df_health_adolescent_res=df_health_adolescent_res.dropna(axis=0,how='all')
df_health_adolescent_res=df_health_adolescent_res.reset_index()
df_health_adolescent_res=df_health_adolescent_res.melt(id_vars=['index'])
# df_health_adolescent_res

#df_health_adolescent_res

#vaccine DPT3, MCV2

df_health_vaccine_sex = df_sdg[df_sdg['Description'].str.contains('diphtheria-tetanus') | df_sdg['Description'].str.contains('vaccine second-dose')]
df_health_vaccine_sex = df_health_vaccine_sex.drop(columns=['Indicator','Series','Unit_Code','Ref_Area','Province','District','Urbanization','Education ','Data Source'])
df_health_vaccine_sex = df_health_vaccine_sex.T
df_health_vaccine_sex.columns = df_health_vaccine_sex.iloc[0]
df_health_vaccine_sex=df_health_vaccine_sex.drop(index=['Description','Age'])
df_health_vaccine_sex.columns=df_health_vaccine_sex.iloc[0]
df_health_vaccine_sex=df_health_vaccine_sex[1:]

#creaating table for DTP3
df_health_vaccine_sex_DTP3 = df_health_vaccine_sex.iloc[::,:3]

#forward fill
df_health_vaccine_sex_DTP3=df_health_vaccine_sex_DTP3.fillna(method='ffill')

#delete rows whose all values are null
df_health_vaccine_sex_DTP3 = df_health_vaccine_sex_DTP3.dropna(axis=0,how='all')
df_health_vaccine_sex_DTP3['indicator'] = 'DTP3'
#df_health_vaccine_sex_DTP3

#create tables for MCV2

df_health_vaccine_sex_MCV2 = df_health_vaccine_sex.iloc[::,3:]

#forward fill
df_health_vaccine_sex_MCV2 = df_health_vaccine_sex_MCV2.fillna(method='ffill')
#delete rows whose all values are null
df_health_vaccine_sex_MCV2 = df_health_vaccine_sex_MCV2.dropna(axis=0,how='all')
df_health_vaccine_sex_MCV2['indicator'] = 'MCV2'

df_health_vaccine=pd.concat([df_health_vaccine_sex_DTP3,df_health_vaccine_sex_MCV2])
#df_health_vaccine


```
